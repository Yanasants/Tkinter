```python
import numpy as np
import matplotlib.pyplot as plt 
from tkinter import *
```

```python
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
class Application:
    def __init__(self, master=None):
        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 20
        self.primeiroContainer.pack()

        self.plotPrim= Button(self.primeiroContainer)
        self.plotPrim["text"] = "Primeiro Grau"
        self.plotPrim["font"] = ("Calibri", "8")
        self.plotPrim["width"] = 12
        self.plotPrim.pack(side=LEFT)
        self.plotPrim["command"] = self.PrimGrau
        self.plotPrim.pack()

        self.plotSeg= Button(self.primeiroContainer)
        self.plotSeg["text"] = "Segundo Grau"
        self.plotSeg["font"] = ("Calibri", "8")
        self.plotSeg["width"] = 12
        self.plotSeg.pack(side=LEFT)
        self.plotSeg["command"] = self.SegGrau
        self.plotSeg.pack()

        self.plotTer= Button(self.primeiroContainer)
        self.plotTer["text"] = "Terceiro Grau"
        self.plotTer["font"] = ("Calibri", "8")
        self.plotTer["width"] = 12
        self.plotTer.pack(side=LEFT)
        self.plotTer["command"] = self.TerGrau
        self.plotTer.pack()
 
    def menu(self):
        self.primeiroContainer.destroy()
        self.segundoContainer.destroy()
        self.terceiroContainer.destroy()
        try:
            self.quartoContainer.destroy()
        except:
            pass
        self.quintoContainer.destroy()
        self.sextoContainer.destroy()
        self.setimoContainer.destroy()

        self.plotando.destroy()
        self.voltando.destroy()

        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame()
        self.primeiroContainer["pady"] = 20
        self.primeiroContainer.pack()

        self.plotPrim= Button(self.primeiroContainer)
        self.plotPrim["text"] = "Primeiro Grau"
        self.plotPrim["font"] = ("Calibri", "8")
        self.plotPrim["width"] = 12
        self.plotPrim.pack(side=LEFT)
        self.plotPrim["command"] = self.PrimGrau
        self.plotPrim.pack()

        self.plotSeg= Button(self.primeiroContainer)
        self.plotSeg["text"] = "Segundo Grau"
        self.plotSeg["font"] = ("Calibri", "8")
        self.plotSeg["width"] = 12
        self.plotSeg.pack(side=LEFT)
        self.plotSeg["command"] = self.SegGrau
        self.plotSeg.pack()

        self.plotTer= Button(self.primeiroContainer)
        self.plotTer["text"] = "Terceiro Grau"
        self.plotTer["font"] = ("Calibri", "8")
        self.plotTer["width"] = 12
        self.plotTer.pack(side=LEFT)
        self.plotTer["command"] = self.TerGrau
        self.plotTer.pack()


    def PrimGrau(self, master=None):
        self.primeiroContainer.destroy()


        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer["padx"] = 20
        self.segundoContainer["pady"] = 10
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer["padx"] = 20
        self.terceiroContainer["pady"] = 10
        self.terceiroContainer.pack()

        self.quintoContainer = Frame(master)
        self.quintoContainer["pady"] = 10
        self.quintoContainer.pack()

        self.sextoContainer = Frame(master)
        self.sextoContainer["pady"] = 30
        self.sextoContainer.pack()

        self.setimoContainer = Frame(master)
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()

        self.titulo = Label(self.primeiroContainer, text="Função do Primeiro Grau")
        self.titulo["font"] = ("Arial", "10", "bold")
        self.titulo.pack()

        self.coefALabel = Label(self.segundoContainer,text="Coeficiente A", font=self.fontePadrao)
        self.coefALabel.pack(side=LEFT)

        self.coefA = Entry(self.segundoContainer)
        self.coefA["width"] = 10
        self.coefA["font"] = self.fontePadrao
        self.coefA.pack(side=LEFT)

        self.coefBLabel = Label(self.terceiroContainer, text="Coeficiente B", font=self.fontePadrao)
        self.coefBLabel.pack(side=LEFT)

        self.coefB = Entry(self.terceiroContainer)
        self.coefB["width"] = 10
        self.coefB["font"] = self.fontePadrao
        self.coefB.pack(side=LEFT)

        self.plotando = Button(self.sextoContainer)
        self.plotando["text"] = "Plotar"
        self.plotando["font"] = ("Calibri", "8")
        self.plotando["width"] = 12
        self.plotando.pack(side=LEFT)
        self.plotando["command"] = self.plotarPrimGrau
        self.plotando.pack()

        self.voltando = Button(self.sextoContainer)
        self.voltando["text"] = "Voltar"
        self.voltando["font"] = ("Calibri", "8")
        self.voltando["width"] = 12
        self.voltando.pack(side=LEFT)
        self.voltando["command"] = self.menu
        self.voltando.pack()

    #Método verificar coefB
    def plotarPrimGrau(self):
        self.setimoContainer.destroy()
        self.setimoContainer = Frame()
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()
        a = float(self.coefA.get())
        b = float(self.coefB.get())

        fig = plt.figure(figsize=(8,6), dpi=80)

        canva = FigureCanvasTkAgg(fig, self.setimoContainer)
        canva.get_tk_widget().pack(side=TOP, fill=BOTH, expand=1)

        x_array= []
        for i in range(-100,100):
            x_array.append(i)
        
        y_array = []
        for i in range(0,len(x_array)): 
            y = a*x_array[i]+b
            y_array.append(y)

        plt.plot(x_array, y_array)  
        plt.show() 

    def SegGrau(self, master=None):
        self.primeiroContainer.destroy()

        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer["padx"] = 20
        self.segundoContainer["pady"] = 10
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer["padx"] = 20
        self.terceiroContainer["pady"] = 10
        self.terceiroContainer.pack()

        self.quartoContainer = Frame(master)
        self.quartoContainer["padx"] = 20
        self.quartoContainer["pady"] = 10
        self.quartoContainer.pack()

        self.quintoContainer = Frame(master)
        self.quintoContainer["pady"] = 10
        self.quintoContainer.pack()

        self.sextoContainer = Frame(master)
        self.sextoContainer["pady"] = 30
        self.sextoContainer.pack()

        self.setimoContainer = Frame(master)
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()

        self.titulo = Label(self.primeiroContainer, text="Função do Segundo Grau")
        self.titulo["font"] = ("Arial", "10", "bold")
        self.titulo.pack()

        self.coefALabel = Label(self.segundoContainer,text="Coeficiente A", font=self.fontePadrao)
        self.coefALabel.pack(side=LEFT)

        self.coefA = Entry(self.segundoContainer)
        self.coefA["width"] = 10
        self.coefA["font"] = self.fontePadrao
        self.coefA.pack(side=LEFT)

        self.coefBLabel = Label(self.terceiroContainer, text="Coeficiente B", font=self.fontePadrao)
        self.coefBLabel.pack(side=LEFT)

        self.coefB = Entry(self.terceiroContainer)
        self.coefB["width"] = 10
        self.coefB["font"] = self.fontePadrao
        self.coefB.pack(side=LEFT)

        self.coefCLabel = Label(self.quartoContainer, text="Coeficiente C", font=self.fontePadrao)
        self.coefCLabel.pack(side=LEFT)

        self.coefC = Entry(self.quartoContainer)
        self.coefC["width"] = 10
        self.coefC["font"] = self.fontePadrao
        self.coefC.pack(side=LEFT)

        self.plotando = Button(self.sextoContainer)
        self.plotando["text"] = "Plotar"
        self.plotando["font"] = ("Calibri", "8")
        self.plotando["width"] = 12
        self.plotando.pack(side=LEFT)
        self.plotando["command"] = self.plotarSegGrau
        self.plotando.pack()

        self.voltando = Button(self.sextoContainer)
        self.voltando["text"] = "Voltar"
        self.voltando["font"] = ("Calibri", "8")
        self.voltando["width"] = 12
        self.voltando.pack(side=RIGHT)
        self.voltando["command"] = self.menu
        self.voltando.pack()

    #Método verificar coefB
    def plotarSegGrau(self):
        self.setimoContainer.destroy()
        self.setimoContainer = Frame()
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()
        a = float(self.coefA.get())
        b = float(self.coefB.get())
        c = float(self.coefC.get())

        fig = plt.figure(figsize=(8,6), dpi=80)

        canva = FigureCanvasTkAgg(fig, self.setimoContainer)
        canva.get_tk_widget().pack(side=TOP, fill=BOTH, expand=1)

        x_array= []
        for i in range(-100,100):
            x_array.append(i)
        
        y_array = []
        for i in range(0,len(x_array)): 
            y = a*x_array[i]**2+x_array[i]*b+c
            y_array.append(y)

        plt.plot(x_array, y_array)  
        plt.show() 
    
    def TerGrau(self, master=None):
        self.primeiroContainer.destroy()


        self.fontePadrao = ("Arial", "10")
        self.primeiroContainer = Frame(master)
        self.primeiroContainer["pady"] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer["padx"] = 20
        self.segundoContainer["pady"] = 10
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer["padx"] = 20
        self.terceiroContainer["pady"] = 10
        self.terceiroContainer.pack()

        self.quartoContainer = Frame(master)
        self.quartoContainer["padx"] = 20
        self.quartoContainer["pady"] = 10
        self.quartoContainer.pack()

        self.quintoContainer = Frame(master)
        self.quintoContainer["padx"] = 20
        self.quintoContainer["pady"] = 10
        self.quintoContainer.pack()

        self.sextoContainer = Frame(master)
        self.sextoContainer["pady"] = 30
        self.sextoContainer.pack()

        self.setimoContainer = Frame(master)
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()

        self.titulo = Label(self.primeiroContainer, text="Função do Terceiro Grau")
        self.titulo["font"] = ("Arial", "10", "bold")
        self.titulo.pack()

        self.coefALabel = Label(self.segundoContainer,text="Coeficiente A", font=self.fontePadrao)
        self.coefALabel.pack(side=LEFT)

        self.coefA = Entry(self.segundoContainer)
        self.coefA["width"] = 10
        self.coefA["font"] = self.fontePadrao
        self.coefA.pack(side=LEFT)

        self.coefBLabel = Label(self.terceiroContainer, text="Coeficiente B", font=self.fontePadrao)
        self.coefBLabel.pack(side=LEFT)

        self.coefB = Entry(self.terceiroContainer)
        self.coefB["width"] = 10
        self.coefB["font"] = self.fontePadrao
        self.coefB.pack(side=LEFT)

        self.coefCLabel = Label(self.quartoContainer, text="Coeficiente C", font=self.fontePadrao)
        self.coefCLabel.pack(side=LEFT)

        self.coefC = Entry(self.quartoContainer)
        self.coefC["width"] = 10
        self.coefC["font"] = self.fontePadrao
        self.coefC.pack(side=LEFT)

        self.coefDLabel = Label(self.quintoContainer, text="Coeficiente D", font=self.fontePadrao)
        self.coefDLabel.pack(side=LEFT)

        self.coefD = Entry(self.quintoContainer)
        self.coefD["width"] = 10
        self.coefD["font"] = self.fontePadrao
        self.coefD.pack(side=LEFT)

        self.plotando = Button(self.sextoContainer)
        self.plotando["text"] = "Plotar"
        self.plotando["font"] = ("Calibri", "8")
        self.plotando["width"] = 12
        self.plotando.pack(side=LEFT)
        self.plotando["command"] = self.plotarTerGrau
        self.plotando.pack()

        self.voltando = Button(self.sextoContainer)
        self.voltando["text"] = "Voltar"
        self.voltando["font"] = ("Calibri", "8")
        self.voltando["width"] = 12
        self.voltando.pack(side=LEFT)
        self.voltando["command"] = self.menu
        self.voltando.pack()

    def plotarTerGrau(self):
        self.setimoContainer.destroy()
        self.setimoContainer = Frame()
        self.setimoContainer["pady"] = 50
        self.setimoContainer.pack()
        a = float(self.coefA.get())
        b = float(self.coefB.get())
        c = float(self.coefC.get())
        d = float(self.coefD.get())

        fig = plt.figure(figsize=(8,6), dpi=80)

        canva = FigureCanvasTkAgg(fig, self.setimoContainer)
        canva.get_tk_widget().pack(side=TOP, fill=BOTH, expand=1)

        x_array= []
        for i in range(-100,100):
            x_array.append(i)
        
        y_array = []
        for i in range(0,len(x_array)): 
            y = a*x_array[i]**3+b*x_array[i]**2+c*x_array[i]+d
            y_array.append(y)

        plt.plot(x_array, y_array)  
        plt.show() 


root = Tk()
root.geometry('600x800')
Application(root)
root.mainloop()
```
